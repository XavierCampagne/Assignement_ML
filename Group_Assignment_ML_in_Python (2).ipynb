```python
#purchaseTime only happens when there is a purchase, so we eliminate the feature.
datatrain = datatrain.drop(['purchaseTime', 'id'], axis = 1)
datatrain.columns
```

```python
datatest = datatest.drop(['label','purchaseTime', 'id'], axis = 1)
datatest.columns
```

```python
#replace the -1 of labels to 0
datatrain = datatrain.replace(-1,0)
datatest = datatest.replace(-1,0)
```

```python
# Separating the  input features (X) and target variable (y)
X = datatrain.drop('label', axis = 1)
y = datatrain['label']
```

```python
from sklearn.model_selection import train_test_split
# split your dataset into training & test parts, X
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

X_temp, y_temp = SMOTE(random_state=42).fit_resample(X_train, y_train)

X_train, X_test, y_train, y_test = train_test_split(X_temp, y_temp)

print(f'''Shape of X before SMOTE: {X.shape}
Shape of X after SMOTE: {X_temp.shape}''')

print('\nBalance of classes in percentage:')
y_temp.value_counts(normalize=True) * 100
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(random_state=42)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
test_score = logmodel.score(X_test, y_test)
print("model score: %.3f " % test_score)
```

```python
y_pred = logmodel.predict(datatest)
print(y_pred)
```

```python
probability = logmodel.predict_proba(datatest)[:,1]
print(probability)
```

```python
data_test = pd.read_csv('test.csv')
result = pd.DataFrame({'id': data_test.id, 'Label' : probability})
result
```
